```python
df_train.columns = ['id', 'age', 'sex', 'chest_pain', 'rest_bps', 'serum_chol', 'fasting_blood_sugar', 'resting_ecg',
       'max_heart_rate', 'exercise_induced_angina', 'oldpeak', 'slope_of_peak', 'no_of_blood_vessels', 'thal', 'target']
df_test.columns = ['id', 'age', 'sex', 'chest_pain', 'rest_bps', 'serum_chol', 'fasting_blood_sugar', 'resting_ecg',
       'max_heart_rate', 'exercise_induced_angina', 'oldpeak', 'slope_of_peak', 'no_of_blood_vessels', 'thal']
```

```python
train = df_train.copy()
```

```python
def age_risk(data):
    if data['sex'] == 0:
        if data['age'] < 45:
            return 'Low'
        elif 45 <= data['age'] <65:
            return 'Moderate'
        else:
            return 'High'

    elif data['sex'] == 1:
        if data['age']< 55:
            return 'Low'
        elif 55 <= data['age'] <65:
            return 'Moderate'
        else:
            return 'High'
    return None
```

```python
train['age_risk_level'] = train.apply(age_risk,axis=1)
```

```python
#Serum Chol feature

def chol_level(chol):
    if chol < 200:
        return 'Okay'
    elif 200 <= chol <=239:
        return 'Borderline_high'
    else:
        return 'High'
```

```python
train['chol_level'] = train.serum_chol.apply(chol_level)
```

```python
def blood_pressure(rbps):
    if rbps <=120:
        return 'Healthy'
    else:
        return 'risk'
```

```python
train['rbps_risk'] = train.rest_bps.apply(blood_pressure)
```

```python
def max_heart_rate(hr):
    if hr <=100:
        return 'healthy'
    else:
        return 'risk'
```

```python
train['heart_rate_risk'] = train['max_heart_rate'].apply(max_heart_rate)
```
